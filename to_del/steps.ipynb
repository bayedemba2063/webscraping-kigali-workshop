### suggested steps for the project

1. create a folder on your computer where you will store all files, data, scripts

2. in vs code set this folder as the project folder (File > Open Folder...)

3. create a virtual environment (use one of the approaches)
    * View > Command Palette... > from the list choose "Python: Create Environment..." "click" and choose 'Venv'


* in terminal use command: python<python_version> -m venv <venv_name> (e.g. python3.10 -m venv scraping)
* e.g. python3.10 -m venv scraping

4. in browser inspect the page you want to scrape (right click > Inspect), look for the patterns in the HTML structure, search for the data you  want to acquire

5. start coding in new python file or jupyter notebook file. In VS Code when you are in your project folder you can do it by:
    * choosing Explorer from the left bar and then in the top of the file list you will see 'create New File'


    * using VS Code top menu, File > New file

6. to create python file remember to use the extension .py, to create jupyter notebook file use .ipynb

7. Remember to activate your virtual environment and always check if you are working on the right one.
    * you can check what is the activated environment by opening New Terminal (Terminal > New Terminal)
    * or it should be visible ine the bottom right corner of VS code

to do it from the Terminal, go to the path where your project is (it should already be visible)
* e.g. C:\Users\domin\PythonProjects\UNECE_training>
* now go to Scripts folder inside the venv folder
* cd scraping\Scripts    <scraping is the name of the virtual environment>
* finally to activate the venv:
* .\activate
